## Lets load Deep learning Model

```python
mobile = tf.keras.applications.mobilenet.MobileNet() #Mobile net is a deep learnign architecture for image classifcation
#Pre trained
```

```python
mobile1 = tf.keras.applications.mobilenet_v2.MobileNetV2()
```

```python
mobile2 = tf.keras.applications.ResNet50()    #mobilenet_v2.MobileNetV2()
```

```python
## Four phases of deep learning
## Creating , Taining , Test or Validate , Predict 
```

## First we need to do preprocessing

```python
from tensorflow.keras.preprocessing import image
img = image.load_img(filename, target_size = (224, 224))
```

```python
plt.imshow(img)
```

```python
resized_img = image.img_to_array(img)
final_image = np.expand_dims(resized_img, axis = 0)## need fourth dimension
final_image = tf.keras.applications.mobilenet.preprocess_input(final_image)
```

```python
resized_img.shape
```

```python
final_image.shape #fourth dimension added
```

## Prediction by using mobile Net version 1

```python
predictions =mobile.predict(final_image) 
```

```python
print(predictions)
```

```python
from tensorflow.keras.applications import imagenet_utils
```

```python
results = imagenet_utils.decode_predictions(predictions)
```

```python
print(results)
```

```python
plt.imshow(img)
```

## Prediction by using mobile Net version 2

```python
predictions1 =mobile1.predict(final_image)
```

```python
results1 = imagenet_utils.decode_predictions(predictions1)
```

```python
print(results1)
```
